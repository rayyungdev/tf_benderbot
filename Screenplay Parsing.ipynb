```python
import requests
import re
from bs4 import BeautifulSoup
from functools import reduce
import numpy as np
from matplotlib import pyplot as plt
from collections import defaultdict
```

Getting screenplay page and split text from webpage.

```python
r = requests.get('https://www.imsdb.com/transcripts/Futurama-Space-Pilot-3000.html')
soup = BeautifulSoup(r.content)
text = soup.find('td', class_='scrtext').get_text()
print(text)
```

Split text into lines

```python
lines = re.split(r'\n+', text)
lines = [
    '' if re.match('^\s+$', line) else line
    for line in lines
]
lines
```

```python
# Start after title
lines = lines[8:]

# Start line iterator
lineiter = iter(lines)

character_regex = re.compile('^\s+([A-Z]+)$')
all_dialogues = defaultdict(list)

# Line iterator will throw us a StopIteration error when we run out of lines
try:
    while True:
        # Find first instance of character
        line = next(lineiter)
        match = character_regex.match(line)
        while not match:
            line = next(lineiter)
            match = character_regex.match(line)
        character = match.group(1)

        # Now find every line until an empty line
        dialogue = []
        line = next(lineiter)
        while line != '':
            dialogue.append(line)
            line = next(lineiter)
        dialogue = [ line.strip() for line in dialogue ]
        dialogue = [ line.rstrip() for line in dialogue ]
        dialogue = ' '.join(dialogue)
        all_dialogues[character].append(dialogue)
    
except StopIteration:
    pass

all_dialogues
```
